```python
import pandas as pd
import numpy as np

csv_file_path = 'PY_STRESS.csv'

df = pd.read_csv(csv_file_path, index_col='No')
```

```python
df.describe
```

```python
USED_STRESS = df[df['Used?'] == 'Yes']
```

```python
USED_STRESS.info
```

```python
import matplotlib.pyplot as plt
```

```python
Software_freq = USED_STRESS['Software'].value_counts()

plt.figure(figsize=(10,8))
SB = Software_freq.plot(kind='bar')

plt.title('Frequency of Software Usage')
plt.xlabel('Software')
plt.ylabel('Frequency')

SB.bar_label(SB.containers[0], label_type='edge', padding=3)


plt.show()
```

```python
Software_method_counts = USED_STRESS.groupby(['Method', 'Software']).size().unstack()

# Plotting the grouped bar chart
plt.figure(figsize=(20, 8))
Software_method_counts.plot(kind='bar',width=3, stacked=False, color=['skyblue', 'orange', 'green', 'red', 'purple'])

# Add titles and labels
plt.title('Frequency of Software Based on Simulation Method')
plt.xlabel('Simulation Method')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Show the legend
plt.legend(title='Software', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
Method_freq = USED_STRESS['Method'].value_counts()

plt.figure(figsize=(10,8))
MB = Method_freq.plot(kind='bar')

plt.title('Frequency of Simulation Methods')
plt.xlabel('Method')
plt.ylabel('Frequency')
# Add data labels on the bars
MB.bar_label(MB.containers[0], label_type='edge', padding=3)


plt.show()
```

```python
USED_STRESS['Year'] = USED_STRESS['Year'].astype(int)

Year_freq = USED_STRESS['Year'].value_counts().sort_index(ascending=False)

plt.figure(figsize=(10,8))
YB = Year_freq.plot(kind='bar')

plt.title('Frequency of Papers over Time')
plt.xlabel('Year')
plt.ylabel('Frequency')

# Add data labels on the bars
YB.bar_label(YB.containers[0], label_type='edge', padding=3)

plt.show()
```
