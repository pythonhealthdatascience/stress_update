```python
import pandas as pd
import numpy as np

csv_file_path = 'PY_STRESS.csv'

df = pd.read_csv(csv_file_path, index_col='No')
```

```python
df.describe
```

```python
USED_STRESS = df[df['Used?'] == 'Yes']
```

```python
USED_STRESS.info
```

```python
import matplotlib.pyplot as plt
```

```python
Software_freq = USED_STRESS['Software'].value_counts()

plt.figure(figsize=(10,8))
SB = Software_freq.plot(kind='bar')

plt.title('Frequency of Software Usage')
plt.xlabel('Software')
plt.ylabel('Frequency')

SB.bar_label(SB.containers[0], label_type='edge', padding=3)


plt.show()
```

```python
Software_method_counts = USED_STRESS.groupby(['Method', 'Software']).size().unstack()

# Plotting the grouped bar chart
plt.figure(figsize=(20, 8))
Software_method_counts.plot(kind='bar',width=3, stacked=False, color=['skyblue', 'orange', 'green', 'red', 'purple'])

# Add titles and labels
plt.title('Frequency of Software Based on Simulation Method')
plt.xlabel('Simulation Method')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Show the legend
plt.legend(title='Software', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
target_authors = [
    'Thomas Monks', 'Christine Currie', 'Bhakti Stephan Onggo', 
    'Stewart Robinson', 'Martin Kunc', 'Simon Taylor'
]
```

```python
USED_STRESS['Authors'] = USED_STRESS['Authors'].astype(str)

# Split authors into separate columns
authors_split = USED_STRESS['Authors'].str.split(',', expand=True).apply(lambda x: x.str.strip())

# Rename columns to reflect the author positions
authors_split.columns = [f'Author_{i+1}' for i in range(authors_split.shape[1])]

# Concatenate with the original DataFrame
USED_STRESS_separated = pd.concat([USED_STRESS, authors_split], axis=1).drop(columns=['Authors'])

```

```python
def contains_target_author(authors_split, target_authors):
    return any(author in target_authors for author in authors_split)
```

```python
# Check if any author is in the target list and create DataFrames accordingly
target_included = USED_STRESS_separated[USED_STRESS_separated['authors_split'].apply(lambda authors: contains_target_author(authors, target_authors))]
not_target_authors = USED_STRESS_separated[~USED_STRESS_separated['authors_split'].apply(lambda authors: contains_target_author(authors, target_authors))]

# Print the resulting DataFrames
print("DataFrame with target authors included:")
print(target_included)

print("\nDataFrame with no target authors:")
print(not_target_authors)

```
